## Задание 1. Анализ данных недвижимости

Условие: 
1. Постройте гистограмму, отображающую распределение цен на недвижимость в наборе данных из файла kc-house-data. 
2. Постройте гистограмму для анализа распределения площади жилых помещений. На графике должно быть ясно видно, какие значения площади наиболее распространены. 
3. Постройте гистограмму, показывающую распределение годов постройки домов. График должен помочь выявить тенденции в строительстве по годам.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Загрузка данных
kc_house_data = pd.read_csv("les/kc-house-data.csv", sep=",")
kc_house_data
```

```python
# Построение графика распределения цен на недвижимость
plt.figure(figsize=(10, 6))
plt.hist(kc_house_data["price"], bins=30, edgecolor="black")
plt.title("Распределение цен на недвижимость")
plt.xlabel("Цена (USD)")
plt.ylabel("Количество домов")
plt.grid(True)
plt.show()
```

На графике заметно, что цена большинство домов в выборке находится в пределах 500 000$

```python
# Построение графика распределения квадратуры жилой площади
plt.figure(figsize=(10, 6))
plt.hist(kc_house_data["sqft_living"], bins=30, edgecolor="black")
plt.title("Распределение жилой площади")
plt.xlabel("Жилая площадь (sqft)")
plt.ylabel("Количество домов")
plt.grid(True)
plt.show()
```

На графике можно увидеть наиболее распространенные размеры жилой площади (~2000 sqft). 

```python
# Построение графика распределения года постройки
plt.figure(figsize=(10, 6))
plt.hist(kc_house_data["yr_built"], bins=30, edgecolor="black")
plt.title("Распределение года постройки")
plt.xlabel("Год постройки")
plt.ylabel("Количество домов")
plt.grid(True)
plt.show()
```

## Задача 2. Анализ характеристик домов

1. Постройте столбчатую диаграмму для анализа количества домов с видом на набережную и без вида. График должен показать соотношение между домами с видом и без него. 
2. Постройте столбчатую диаграмму, отображающую распределение домов по количеству этажей. График должен наглядно показать, как часто встречаются дома с разным количеством этажей. 
3. Постройте столбчатую диаграмму для анализа распределения состояния домов. График должен показывать частоту встречаемости различных состояний домов.

```python
# Построение графика распределения домов по наличию вида на набережную
plt.figure(figsize=(5, 5))
kc_house_data["waterfront"].value_counts().plot(kind="bar", color=["skyblue", "salmon"])
plt.hist(kc_house_data["yr_built"], bins=30, edgecolor="black")
plt.title("Распределение домов по наличию вида на набережную")
plt.xlabel("Вид на набережную)")

plt.title("Распределение домов по наличию вида на набережную")
plt.ylabel("Количество домов")
plt.xticks([0, 1], ["Нет", "Да"], rotation=0)

plt.grid(axis="y")
plt.show()
```

На графике можно определить, чтоб большинство домов не имеют вид на набережную и как это соотносится с общим количеством домов

```python
# Построение графика распределения этажей
plt.figure(figsize=(8, 5))
kc_house_data["floors"].value_counts().sort_index().plot(kind="bar", color="lightgreen")

plt.title("Распределение этажей домов")

plt.xlabel("Количество этажей)")
plt.ylabel("Количество домов")

plt.grid(axis="y")
plt.show()
```

## Задача 3. Анализ влияния характеристик на стоимость недвижимости

1. Постройте диаграмму рассеяния, чтобы изучить зависимость между количеством спален и стоимостью недвижимости. График должен показать, как изменение количества спален влияет на цену. 
2. Постройте график размаха (boxplot) для анализа влияния состояния дома на его стоимость. График должен продемонстрировать, как состояние дома связано с его ценой и выявить возможные отклонения. 
3. Постройте график размаха (boxplot), чтобы исследовать, как количество этажей влияет на стоимость недвижимости. График должен показать, как ценовые данные варьируются в зависимости от этажности. 
4. Постройте график размаха (boxplot), чтобы проанализировать, как наличие вида на набережную влияет на стоимость недвижимости. График должен показать, есть ли значительное влияние наличия вида на цену. 
5. Постройте диаграмму рассеяния, чтобы исследовать зависимость между площадью жилой зоны и стоимостью недвижимости. График должен показать, как площадь жилой зоны коррелирует с ценой.

```python
# Построение графика распределения цен в зависимости от количества спален.
plt.figure(figsize=(12, 6))
plt.scatter(kc_house_data["bedrooms"], kc_house_data["price"], alpha=0.5, color="blue")

plt.title("Цена недвижимости в зависимости от количества спален")

plt.xlabel("Количество спален")
plt.ylabel("Цена (USD)")

plt.grid(True)
plt.show()
```

```python
# Построение графика распределения цен в зависимости от состояния.
plt.figure(figsize=(12, 6))
plt.boxplot(
    [
        kc_house_data[kc_house_data["condition"] == i]["price"]
        for i in sorted(kc_house_data["condition"].unique())
    ],
    tick_labels=sorted(kc_house_data["condition"].unique()),
)

plt.title("Цена недвижимости в зависимости от состояния")

plt.xlabel("Состояние дома")
plt.ylabel("Цена (USD)")

plt.grid(True)
plt.show();
```

```python
# Построение графика распределения цен в зависимости от этажности.
plt.figure(figsize=(12, 6))
plt.boxplot(
    [
        kc_house_data[kc_house_data["floors"] == i]["price"]
        for i in sorted(kc_house_data["floors"].unique())
    ],
    tick_labels=sorted(kc_house_data["floors"].unique()),
)

plt.title("Цена недвижимости в зависимости от этажности")

plt.xlabel("Количество этажей")
plt.ylabel("Цена (USD)")

plt.grid(True)
plt.show()
```

```python
# Построение графика распределения цен в зависимости от вида на набережную.
plt.figure(figsize=(12, 6))
kc_house_data.boxplot(column="price", by="waterfront", grid=False)

plt.title("Цена недвижимости в зависимости от наличия вида на набережную")
plt.suptitle("")  # Удаление дополнительного заголовка, добавляемого pandas
plt.xlabel("Вид на набережную")
plt.ylabel("Цена (USD)")
plt.xticks([1, 2], ["Нет", "Да"])

plt.grid(True)
plt.show()
```

```python
# Построение графика распределения цен в зависимости от площади жилой зоны.
plt.figure(figsize=(12, 6))
plt.scatter(
    kc_house_data["sqft_living"], kc_house_data["price"], alpha=0.5, color="green"
)

plt.title("Цена недвижимости в зависимости от площади жилой зоны")
plt.xlabel("Жилая площадь (sqft")
plt.ylabel("Цена (USD)")

plt.grid(True)
plt.show()
```
